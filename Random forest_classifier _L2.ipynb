```python
L2_op = pd.read_csv('/content/drive/MyDrive/Airlines project/df_L2.csv')
L2_op.head()
```

Read the file path using pandas library to import into the colab notebook from mounted drive

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
L2_op ["L2_OP_PER"]=lb.fit_transform(L2_op["L2_OP_PER"])
L2_op.head()
```

As a part of data preprocessing, "L2_op_per" column have multi classes,  we have labelencoded using library LabelEncoder from sklearn kit

```python
# Input and Output Split
predictors = L2_op.loc[:, L2_op.columns!="L2_OP_PER"]
type(predictors)
```

Here, splitting or dividing the dataset into predictors and target variable for machine learning model building.

```python
target = L2_op["L2_OP_PER"]
type(target)  
```

created a target variable for L2-op-per

```python
# Train Test partition of the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.3, random_state= 42)
```

splitting the data into train and test data using predictor variables and target variable to train the model 

```python
from sklearn.ensemble import RandomForestClassifier

L2_op_clf = RandomForestClassifier(n_estimators=500, n_jobs=1, random_state=42)
```

imported the Randomforest classifier from sklearn kit, for model building added with parameters required for this model,applied random_state to iterate the data for 42 times

```python
L2_op_clf.fit(x_train, y_train)
```

Fit the model under X-train and y-train 

```python
from sklearn.metrics import accuracy_score, confusion_matrix
confusion_matrix(y_test, L2_op_clf.predict(x_test))
```

performance of the model evaluated using confusion matrix, imported confusion matrix and accuracy score from sklearn metrics and found false output observed in confusion matrix

```python
accuracy_score(y_test, L2_op_clf.predict(x_test))
```

accuracy score was captured with highest of 0.999996
